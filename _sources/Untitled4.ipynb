## Korrektive tiltake mot buling ##

Vi har bulespenningsformelen gitt ved: 
$$ \sigma_{Euler} = \frac{ C }{1-\nu^2 } \times \frac{\pi^2 \times E}{12} \times (\frac{t}{s})^2$$

Vi deler denne formelen gjerne i 3 deler: 

### Bulekoffefisienten C  ###
 
 C vil kunne kes fra typisk 1.25 til 4 ved  snu avstivningsretningen! (Ufordelaktig  ha trykkspenning langs langsiden) 
 Dette er ganske vanlig p skipskonstruksjoner  da la avstivningsretningen vre langskips.  
 




 
 
 ### Materialkonstantenten  ###

$$ \frac{\pi^2 \times E}{ ({1-\nu^2) } \times 12}$$ 
 
For stl 189800, og kun avhengig av E-modul. Ingen betydning i flytspenningen. 

 ### Slankheten  ###
 
 $$ (\frac{t}{s})^2$$
 
 Avhengig av platetykkelsen og korteste avstand til avstivningselementer.
 - En kning i platetykkelse vil som regel vre meget vektintensivt, i forhold til  minske stiveravstanden s. 
 - Det motsatte gjelder nr produksjonsvennlighet er nskelig.
