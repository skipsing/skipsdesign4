# Platebuling 

#### Repetisjon fra bjelkeknekking ####
Vi har fra tidligere at kritisk last p en aksialt belastet bjelke er definert som: $$P_{kr}= \frac{\pi^2 \times E \times I }{L^2}$$

Vi skal n se p tilsvarende knekkingsfenomen p platefelt. Dette er typisk for skrogkonstruksjoner bygd opp av *paneler*. Hvert panel er bygd opp av en plate som er psveist ulike avstivningelementer (*stivere og brere*). 


Betrakter en platestripe med bredde $b$ markert i rdt over som en syle og anvender knekklastformelen for $P_{kr}$:

$$ P_{kr} = \frac{\pi^2 \times E \times (\frac{b \times t^3}{12}) }{s^2} $$ <br>

Annet arealmoment for det sammensatte bjelkesnittet ($I$) er her erstattet med annet arealmoment for en rektangulr plate $(\frac{b \times t^3}{12})$ der hyde er $b$ og tykkelse er $t$. Videre er lengden $L$ erstattet med $s$, 
se figur under. <br>  

Eulerspenningen beregnes videre som knekklast delt p areal: $$ \sigma_{Euler} = \frac{P_{Kr}}{A} = \frac{\pi^2 \times E \times (\frac{b \times t^3}{12}) }{(b \times t) \times s^2} $$ <br> $$\downarrow $$ <br> $$\sigma_{Euler} = \frac{\pi^2 \times E}{12} \times (\frac{t}{s})^2$$

For tilfellene der platen er opplagret langs kantene (som den ofte er) s korrigeres det p samme mte som for bjelker med effektive knekklengder, men her inkluderes ogs spenningsvariasjon over platefeltet. 

$$\frac{ C }{1-\nu^2 }$$

Der $C$ er en koeffisient som sier noe om platefeltets aspektforhold og spenningsbilde. Denne er ofte ogs benevnt *bulekoeffisienten*. $\nu$ er *Poisson's ratio*, eller *tverrkontrakssjonstallet*. For stl er $\nu = 0.3 $.

Under er tabell for bulekoeffisienten for ulike spenningsvariasjoner og spenningsretninger angitt: 


Vi str da igjen med flgende uttrykk for Eulerspenning i et vilkrlig platefelt:
<br>
<br>
$$ \sigma_{Euler} = \frac{ C }{1-\nu^2 } \times \frac{\pi^2 \times E}{12} \times (\frac{t}{s})^2$$
<br>
<br>

Vi husker fra bjelkeknekkingen at det ble korrigert for tilvirkningsmetode for  ta hensyn til imperfeksjoner. 
<br>For platebuling er dette utrykt ved *Johnson-Ostenfeld korreksjonsfaktor*: 



En ser fra figur over at denne korreksjonen inntreffer om $\sigma_{Euler} > 0.5\times \sigma_{flyt}
$.Det gir flgende korreksjon:

$$ \sigma_{kr} = \sigma_{Euler}  \textrm{  nr  } \sigma_{Euler} \le  \frac{\sigma_{flyt}}{2}   $$
<br>
$$ \sigma_{kr} = \sigma_{flyt} \times (1-\frac{\sigma_{flyt}}{4 \times\sigma_{Euler} })  \textrm{  nr  } \sigma_{Euler} \ge  \frac{\sigma_{flyt}}{2}   $$

### Regneeksempel ###

Vi skal kontrollere platefeltet under for buling. Platefeltet er belastet med en varierende trykkspenning langs kortsiden. 
<br>

$$E_{modul} = 210 000 \frac{N}{mm^2} $$
<br>
$$\sigma_{flyt} = 235 \frac{N}{mm^2} $$

#### Prosedyre ####
- Identifiser spenningstilfelle fra bulekoeffisientarket. 
- Beregn spenningsvariasjonen $\psi$.  
- Beregn tilhrende bulekoeffisient $C$. 
- Beregn $\sigma_{Euler}$ gitt aspektforhold ($\frac{t}{s}$) og materialparameterene ( $E$ og $\nu$ ) 
- korriger $\sigma_{kr}$ for Johnson Ostenfeld korreksjonsfaktoren
- platen vil bule om $\sigma_{kr} < \sigma_{max}$ 

#### Lsningsforslag ####

En ser her at dette er et tilfelle a), og der $\sigma_{2} = 0\frac{N}{mm^2}$, og $\psi$ blir da ogs $0$.  

Bulekoeffisienten blir $C = \frac{8.4}{0 + 1.1} = 7.63$
<br>
<br>
Beregner Eulerspenningen; $ \sigma_{Euler} = \frac{ 7.63 }{1-0.3^2 } \times \frac{\pi^2 \times 210 000\frac{N}{mm^2}}{12} \times (\frac{8mm}{600mm})^2 = 257.4\frac{N}{mm^2} $
<br>
<br>
En ser her at $\sigma_{Euler} > 0.5\times \sigma_{flyt}$ og m derfor korrigere for Johnson Ostenfeld:
<br>
<br>
Kritisk bulespenning beregnes da som: $ \sigma_{kr} = 235\frac{N}{mm^2} \times (1-\frac{235\frac{N}{mm^2}}{4 \times 257.4\frac{N}{mm^2} })=234.7\frac{N}{mm^2} $

Kontroll opp mot opptredende spenning: $\sigma_{kr} > \sigma_{max}$ -> OK!!
