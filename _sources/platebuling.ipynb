# Platebuling 

#### Repetisjon fra bjelkeknekking ####
Vi har fra tidligere at kritisk last på en aksialt belastet bjelke er definert som: $$P_{kr}= \frac{\pi^2 \times E \times I }{L^2}$$

Vi skal nå se på tilsvarende knekkingsfenomen på platefelt. Dette er typisk for skrogkonstruksjoner bygd opp av *paneler*. Hvert panel er bygd opp av en plate som er påsveist ulike avstivningelementer (*stivere og bærere*). 


Betrakter en platestripe med bredde $b$ markert i rødt over som en søyle og anvender knekklastformelen for $P_{kr}$:

$$ P_{kr} = \frac{\pi^2 \times E \times (\frac{b \times t^3}{12}) }{s^2} $$ <br>

Annet arealmoment for det sammensatte bjelkesnittet ($I$) er her erstattet med annet arealmoment for en rektangulær plate $(\frac{b \times t^3}{12})$ der høyde er $b$ og tykkelse er $t$. Videre er lengden $L$ erstattet med $s$, 
se figur under. <br>  

Eulerspenningen beregnes videre som knekklast delt på areal: $$ \sigma_{Euler} = \frac{P_{Kr}}{A} = \frac{\pi^2 \times E \times (\frac{b \times t^3}{12}) }{(b \times t) \times s^2} $$ <br> $$\downarrow $$ <br> $$\sigma_{Euler} = \frac{\pi^2 \times E}{12} \times (\frac{t}{s})^2$$

For tilfellene der platen er opplagret langs kantene (som den ofte er) så korrigeres det på samme måte som for bjelker med effektive knekklengder, men her inkluderes også spenningsvariasjon over platefeltet. 

$$\frac{ C }{1-\nu^2 }$$

Der $C$ er en koeffisient som sier noe om platefeltets aspektforhold og spenningsbilde. Denne er ofte også benevnt *bulekoeffisienten*. $\nu$ er *Poisson's ratio*, eller *tverrkontrakssjonstallet*. For stål er $\nu = 0.3 $.

Under er tabell for bulekoeffisienten for ulike spenningsvariasjoner og spenningsretninger angitt: 


Vi står da igjen med følgende uttrykk for Eulerspenning i et vilkårlig platefelt:
<br>
<br>
$$ \sigma_{Euler} = \frac{ C }{1-\nu^2 } \times \frac{\pi^2 \times E}{12} \times (\frac{t}{s})^2$$
<br>
<br>

Vi husker fra bjelkeknekkingen at det ble korrigert for tilvirkningsmetode for å ta hensyn til imperfeksjoner. 
<br>For platebuling er dette utrykt ved *Johnson-Ostenfeld korreksjonsfaktor*: 



En ser fra figur over at denne korreksjonen inntreffer om $\sigma_{Euler} > 0.5\times \sigma_{flyt}
$.Det gir følgende korreksjon:

$$ \sigma_{kr} = \sigma_{Euler}  \textrm{  når  } \sigma_{Euler} \le  \frac{\sigma_{flyt}}{2}   $$
<br>
$$ \sigma_{kr} = \sigma_{flyt} \times (1-\frac{\sigma_{flyt}}{4 \times\sigma_{Euler} })  \textrm{  når  } \sigma_{Euler} \ge  \frac{\sigma_{flyt}}{2}   $$

### Regneeksempel ###

Vi skal kontrollere platefeltet under for buling. Platefeltet er belastet med en varierende trykkspenning langs kortsiden. 
<br>

$$E_{modul} = 210 000 \frac{N}{mm^2} $$
<br>
$$\sigma_{flyt} = 235 \frac{N}{mm^2} $$

#### Prosedyre ####
- Identifiser spenningstilfelle fra bulekoeffisientarket. 
- Beregn spenningsvariasjonen $\psi$.  
- Beregn tilhørende bulekoeffisient $C$. 
- Beregn $\sigma_{Euler}$ gitt aspektforhold ($\frac{t}{s}$) og materialparameterene ( $E$ og $\nu$ ) 
- korriger $\sigma_{kr}$ for Johnson Ostenfeld korreksjonsfaktoren
- platen vil bule om $\sigma_{kr} < \sigma_{max}$ 

#### Løsningsforslag ####

En ser her at dette er et tilfelle a), og der $\sigma_{2} = 0\frac{N}{mm^2}$, og $\psi$ blir da også $0$.  

Bulekoeffisienten blir $C = \frac{8.4}{0 + 1.1} = 7.63$
<br>
<br>
Beregner Eulerspenningen; $ \sigma_{Euler} = \frac{ 7.63 }{1-0.3^2 } \times \frac{\pi^2 \times 210 000\frac{N}{mm^2}}{12} \times (\frac{8mm}{600mm})^2 = 257.4\frac{N}{mm^2} $
<br>
<br>
En ser her at $\sigma_{Euler} > 0.5\times \sigma_{flyt}$ og må derfor korrigere for Johnson Ostenfeld:
<br>
<br>
Kritisk bulespenning beregnes da som: $ \sigma_{kr} = 235\frac{N}{mm^2} \times (1-\frac{235\frac{N}{mm^2}}{4 \times 257.4\frac{N}{mm^2} })=234.7\frac{N}{mm^2} $

Kontroll opp mot opptredende spenning: $\sigma_{kr} > \sigma_{max}$ -> OK!!
