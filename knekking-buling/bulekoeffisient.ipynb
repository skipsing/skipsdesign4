```python

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
# Itererbart intervall for psi fra -1 til 1
psi = np.linspace(-2 , 1 , 100 )

x_verdier = []
y_verdier = []

for n in psi: 
    # Ved strekk 
    if n < 0:
        bulekoeffisient = 7.6 - 6.4 * n + 10 * n ** 2
    # Ingen strekk
    else:
        bulekoeffisient = 8.4 / ( n + 1.1 )
    # Lagrer i lister for plot
    x_verdier.append(n)
    y_verdier.append(bulekoeffisient)



# Plotter

plt.suptitle('Bulekoeffisient som funksjon av ψ')
plt.ylabel('C [-]')
plt.xlabel('ψ [-]')
plt.plot(x_verdier , y_verdier)
plt.show()

```

## KLARIFISERING: Beregning a bulekoeffisient $C$ 

Det er en feil i løsningsforslag til eksamen Høst 2021 Oppgave 4b (s.15)
<br>


Det er oppgitt at strekkspenning i nedre flens er $\sigma_{nedre} = 188MPa $, mens trykkspenningen i øvre flens er $\sigma_{øvre} = 99.4MPa $. <br> Feilen kommer nå i valg av $\sigma_1$ og $\sigma_2$ ut fra formelarket for bulekoeffisienten: <br>
### $\sigma_1$ skal alltid være *største* trykkspenning som er beregnet. 
<br>
Siden $\sigma_{øvre}$ er største (og eneste trykkspenning av $\sigma_{øvre}$ og $\sigma_{nedre}$) så skal ligningen være slik: <br> <br>
<br>
$$ \sigma_{nedre}  = \psi \times \sigma_{øvre} $$ 
<br>
$$ \downarrow $$
<br>
$$ -188MPA  = \psi \times 99.4MPa $$ 
<br>
$$ \psi = -1.89$$
<br>
NB: Trykk og strekk er motsatt rettet og en av de må derfor defineres som negativ: her valgt strekk om negativ (-)
