# Plotting av *bulekoeffisienten* $C$ 
I denne notebooken skal vi beregne og plotte bulekoeffisienten $C$ for trykkspenninger parallelt med plateretningen og på tverrs av plateretningen,   
Vi skal her benytte oss av to biblioteker som ikke er del av standardbiblioteket til Python: 
- Numpy. Denne skal hjelpe oss å lage et intervall av punkter for $\psi$
- Matplotlib. Standard plottebibliotek som skal hjelpe oss å lage presentable plot av $C$


```python
import numpy as np
import matplotlib.pyplot as plt
```

## Spenningsretning langs plateretningen
Vi definerer her spenningsretningen som virker på kortsiden av platen og har retning parallelt med lengste kant som "langs plateretningen"

I bulekoeffisientarket svarer dette til tilfelle a), b) og c). <br>

NB! Vi ser her bort fra tilfelle b) da tilfelle c) gir samme verdi for $ \psi = -1 $

# Sett inn link til bulekoeffsientarket

https://cdn.jsdelivr.net/gh/skipsing/skipsdesign4/images/kraftgang-med-forklaring.PNG

```python

%matplotlib inline

# Itererbart intervall for psi fra -1 til 1
psi_interval = np.linspace(-1 , 1 , 100 )

x_verdier = []
y_verdier = []

for psi in psi_interval: 
    # Ved strekk 
    if psi < 0:
        bulekoeffisient = 7.6 - 6.4 * psi + 10 * psi ** 2
    # Ingen strekk
    else:
        bulekoeffisient = 8.4 / ( psi + 1.1 )
    # Lagrer i lister for plot
    x_verdier.append(psi) 
    y_verdier.append(bulekoeffisient)



# Plotter

plt.suptitle('Bulekoeffisient som funksjon av ψ')
plt.ylabel('C [-]')
plt.xlabel('ψ [-]')
plt.plot(x_verdier , y_verdier)
plt.show()

```

## KLARIFISERING: Beregning a bulekoeffisient $C$ 

Det er en feil i løsningsforslag til eksamen Høst 2021 Oppgave 4b (s.15)
<br>


Det er oppgitt at strekkspenning i nedre flens er $\sigma_{nedre} = 188MPa $, mens trykkspenningen i øvre flens er $\sigma_{øvre} = 99.4MPa $. <br> Feilen kommer nå i valg av $\sigma_1$ og $\sigma_2$ ut fra formelarket for bulekoeffisienten: <br>
### $\sigma_1$ skal alltid være *største* trykkspenning som er beregnet. 
<br>
Siden $\sigma_{øvre}$ er største (og eneste trykkspenning av $\sigma_{øvre}$ og $\sigma_{nedre}$) så skal ligningen være slik: <br> <br>
<br>
$$ \sigma_{nedre}  = \psi \times \sigma_{øvre} $$ 
<br>
$$ \downarrow $$
<br>
$$ -188MPA  = \psi \times 99.4MPa $$ 
<br>
$$ \psi = -1.89$$
<br>
NB: Trykk og strekk er motsatt rettet og en av de må derfor defineres som negativ: her valgt strekk om negativ (-)
